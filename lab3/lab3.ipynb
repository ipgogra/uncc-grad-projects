### 1) Load the Diabetes dataset into a dataframe (10 Marks)

```python


from sklearn.datasets import load_diabetes

diabetes_data=load_diabetes()
##print(diabetes_data.DESCR)

```

```python
import pandas as pd

diabetes_df=pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)
diabetes_df['Output']=diabetes_data.target
diabetes_df
```

### 2) Create your own Training and Testing data and print them out (10 Marks)

```python
from sklearn.model_selection import train_test_split

X = diabetes_df.drop(['Output'], axis=1)
Y = diabetes_df['Output']

x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

### 3) Fit a model using a single feature. Use 'BMI' as the selected feature (10 Marks)

```python
from sklearn import linear_model

model = linear_model.LinearRegression()

model.fit(x_train['bmi'].values.reshape(-1, 1), y_train)
```

### 4) Predict the data, print it and compare it with the original (10 Marks)

```python
y_pred = model.predict(x_test['bmi'].values.reshape(-1,1))
```

```python


y_pred = pd.DataFrame(y_pred)

y_pred
```

```python
y_test,y_pred
```

### 5) Plot the data (10 Marks)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# label axes 
plt.xlabel(" BMI ") 
plt.ylabel(" Median ")

# Scatter plot

plt.scatter(x_test['bmi'], y_test, color ='black')

# line plot for predictions 
plt.plot(x_test['bmi'], y_pred, color = "blue", linewidth=3)

##plt.xticks(())
##plt.xticks(())

##plt.show()

### results in errror: "KeyError: (slice(None, None, None), None)"?
```

### 6) Print the Model Performance (10 Marks)

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(y_test, y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(y_test, y_pred))
```
